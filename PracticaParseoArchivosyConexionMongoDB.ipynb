```python
import re
import pandas as pd
```

```python
regexFrom = re.compile(r'From: (.*@.*)\n')
regexTo = re.compile(r'To: (.*@.*)\n')
regexSubject = re.compile(r'Subject: \[sakai\] svn commit: (r\d{5})')
regexDateCorreo = re.compile(r'Date: ([A-Z][a-z]{2}, [0-9]{1,2} [A-Z][a-z]{2} [0-9]{4})')
regexDateCommit = re.compile(r'Date: (\d{4}-\d{2}-\d{2})')
```

```python
archivo = open("/content/drive/MyDrive/Herramientas3/datosmail.txt") # Esta función nos permite abrir y leer archivos
lineas = []
listaFrom = []
listaTo = []
listaSub = []
listaDateCorreo = []
listaDateCommit = []
for linea in archivo:
  if linea.startswith("From:"):
    r = regexFrom.search(linea)
    if r:
      listaFrom.append(r.group(1))
  if linea.startswith("To:"):
    r = regexTo.search(linea)
    if r:
      listaTo.append(r.group(1))
  if linea.startswith("Subject:"):
    r = regexSubject.search(linea)
    if r:
      listaSub.append(r.group(1))
  if linea.startswith("Date:"):
    r = regexDateCommit.search(linea)
    if r:
      listaDateCommit.append(r.group(1))
    r = regexDateCorreo.search(linea)
    if r:
      listaDateCorreo.append(r.group(1))
      
```

```python
dfLogCorreos = pd.DataFrame(zip(listaFrom,
listaTo,
listaSub,
listaDateCorreo,
listaDateCommit))
```

```python
dfLogCorreos.rename(columns={0:"From",1:"To",2:"Subject",3:"DateCorreo",4:"DateCommit"},
                    inplace=True)
```
